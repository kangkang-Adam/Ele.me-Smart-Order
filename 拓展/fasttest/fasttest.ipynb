```python
import fasttext, os, time

data  = 'train.txt'
model = 'cls.bin'          # 分类模型
# model = 'vec.bin'        # 若做词向量改成这条

start = time.time()
if os.path.exists(model):
    os.remove(model)

# ① 文本分类
model = fasttext.train_supervised(
        input=data,
        epoch=25,
        lr=0.5,
        wordNgrams=2,
        dim=100,
        loss='softmax')
model.save_model(model)

# ② 若只想训词向量，用下面这行代替上面整段
# model = fasttext.train_unsupervised(input=data, model='skipgram', dim=100, epoch=25)
# model.save_model(model)

print('训练完成，耗时 {:.1f}s，模型大小 {:.2f}MB'.format(
        time.time()-start, os.path.getsize(model)/1024/1024))
```
